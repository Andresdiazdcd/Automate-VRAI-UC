```python
import pandas as pd

df = pd.read_excel("Cursos corregidos 20261_vf.xlsx", sheet_name="output")

```

```python
print(df.columns)
```

```python

df['Escuela'] = (
    df['Escuela']
    .astype(str)
    .str.replace(r'^\d+\s*-\s*', '', regex=True)
    .str.strip()
)
```

```python
print(df[['Escuela']].head(10))
```

```python
with pd.ExcelWriter(
    "Cursos corregidos 20261_vf.xlsx",
    engine="openpyxl",
    mode="a",
    if_sheet_exists="replace"
) as writer:
    df.to_excel(writer, sheet_name="output", index=False)

```
